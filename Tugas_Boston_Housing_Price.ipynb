**Tugas Persoalan Regresi**

> Dataset berasal dari *Boston Housing Price*. Berikut adalah proses membuka dataset tersebut sekaligus memisahkan antara data untuk training dan testing. Dari kode di bawah, diketahui bahwa terdapat 404 data dengan 13 atribut dari dataset, sesuai dengan [link berikut](http://lib.stat.cmu.edu/datasets/boston).



> Setelah itu, dilakukan pengamatan terhadap struktur dari data. Data training dan testing memiliki 13 input atau fitur berupa nilai kontinu dan sebuah output yang juga kontinu. Untuk itu, Neural network yang akan dibangun akan menyelesaikan masalah **regresi**.







```python
#Library dataset
from keras.datasets import boston_housing

#Proses memilah data training dan testing
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()

#Proses melihat bentuk dan contoh input-output
print(train_data.shape[1])
print(train_data[2])
print(train_targets[2])
```

**Proses membangun model NN**

Selanjutnya, dibuat arsitektur atau model dari Neural Network yang digunakan. Gambar di bawah adalah arsitektur terpilih dari beberapa percobaan yang telah dilakukan. Berikut adalah lapisan-lapisan (layers) yang digunakan:

1.   Input layer
2.   Layer untuk normalisasi
3.   Hidden layer 1 (128 nodes)
4.   Hidden layer 2 (128 nodes)
5.   Hidden layer 3 (128 nodes)
6.   Hidden layer 4 (128 nodes)
7.   Output layer

Berikut adalah ilustrasi dari arsitektur FCNN yang digunakan.

![](https://drive.google.com/uc?export=view&id=1HAo4wXvezAAh8rJgB3bxvywgHmK4hJ49)

```python
#Import library keras untuk layer-layer dari neural network
import keras
from keras import layers

#Layer normalisasi untuk normalisasi input berdasarkan data training (train_data)
normalizer = layers.Normalization()
normalizer.adapt(train_data)

#Membangun model neural network
model = keras.Sequential(
    [
        #Layer input
        layers.Input(shape=train_data.shape[1]),
     
        #Normalisasi
        normalizer,
        
        #Hidden layer 1
        layers.Dense(128, activation="relu"),
        
        #Hidden layer 2
        layers.Dense(128, activation="relu"),
        
        #Hidden layer 3
        layers.Dense(128, activation="relu"),
        
        #Hidden layer 4
        layers.Dense(128, activation="relu"),
              
        #Output
        layers.Dense(1)
    ]
)

model.summary()
```

**Pemilihan training options**

Berikut adalah proses kompilasi model yang telah dibuat sekaligus memilih beberapa parameter untuk proses training sesuai dengan laporan:


*   Optimizer : Adam
*   Loss function : Mean squared error




```python
#Proses kompilasi model beserta beberapa parameter training
model.compile(loss='MeanSquaredError', optimizer='adam', metrics=['mae'])
```

**Proses training**

Berikut adalah proses training beserta pemilihan beberapa parameter untuk proses training:


*   Epoch
*   Batch size



```python
#Library waktu
from time import time

#Variabel penyimpan waktu mulai trainig
start = time()

#Proses training
model.fit(train_data, train_targets, epochs=200, batch_size=32)

#Perhitungan waktu training
print(time()-start)

#Menghitung MAE score saat model diterapkan pada data training
train_mse_score, train_mae_score = model.evaluate(train_data, train_targets)
print(train_mae_score)
```

**Proses Evaluasi**

Selanjutnya, dilakukan evaluasi model yang telah dibangun dan dilatih. Pertama-tama, dilakukan evaluasi terhadap data testing. Selanjutnya, dilakukan beberapa prediksi pada data testing dan membandingkan hasilnya dengan target di data testing.

Data ke-3 adalah contoh hasil regresi yang relatif jauh dari target. Data ke-4 adalah contoh hasil regresi yang relatif dekat dari target.

```python
#Menghitung MAE score saat model diterapkan pada data testing
test_mse_score, test_mae_score = model.evaluate(test_data, test_targets)
print(test_mae_score)

#Melakukan prediksi terhadap beberapa data testing dan membandingkan dengan target di data testing
for i in range(10):
  predictions = model.predict(test_data[i],)
  print("Data ke-{}".format(i))
  print("Input:")
  print(test_data[i])
  print("Output di dataset:")
  print(test_targets[i])
  print("Output hasil regresi:")
  print(predictions)
  print("______________________")

```
